```python
file_path = r'C:\Users\halla\OneDrive\Documents\GitHub\property_validation_package\Excel_transfer_try.csv'
headers = ['P_val', 'T_val', 'h_val_CEA', 's_val_CEA', 'k_val_CEA', 'mu_val_CEA', 'Cp_val_CEA', 'gam_val_CEA', 'rho_val_CEA', 'R_val_CEA', 'h_val_pkg', 's_val_pkg', 'k_val_pkg', 'mu_val_pkg', 'Cp_val_pkg', 'gam_val_pkg', 'rho_val_pkg', 'R_val_pkg', 'h_val_err', 's_val_err', 'k_val_err', 'mu_val_err', 'Cp_val_err', 'gam_val_err', 'rho_val_err', 'R_val_err', 'outlier_val']
data = pd.read_csv(file_path, usecols=headers, sep=',')

def twodplot(data, xdata,  ydata, title = None, xlabel = None, ylabel = None, colors = ['b', 'g', 'r', 'k'], markers=['.', 'o', 'x'], numberofgraphs=1):
    if numberofgraphs==1:
        for i in range(len(ydata)):
            plot = plt.plot(data[xdata[0]],
                            data[ydata[i]],
                            label = ydata[i],
                            color = colors[i],
                            marker=markers[i])
            plt.title(ydata[0]+' vs '+xdata[0])
            plt.xlabel(xdata[0])
            plt.ylabel(ydata[0])
            plt.legend()
    elif numberofgraphs==2:
        plot1 = plt.subplot(121)
        plot1.plot(data[xdata[0]],
                        data[ydata[0]],
                        label = ydata[0],
                        color = colors[0],
                        marker=markers[0])
        plot1.set_title(ydata[0]+' vs '+xdata[0])
        plot1.set_xlabel(xdata[0])
        plot1.set_ylabel(ydata[0])
        plot2 = plt.subplot(122)
        plot2.plot(data[xdata[0]],
                        data[ydata[1]],
                        label = ydata[1],
                        color = colors[1],
                        marker=markers[1])
        plot2.set_title(ydata[1]+' vs '+xdata[0])
        plot2.set_xlabel(xdata[0])
        plot2.set_ylabel(ydata[1])
        plt.tight_layout()
    elif numberofgraphs==3:
        plot1 = plt.subplot(221)
        plot1.plot(data[xdata[0]],
                        data[ydata[0]],
                        label = ydata[0],
                        color = colors[0],
                        marker=markers[0])
        plot1.set_title(ydata[0]+' vs '+xdata[0])
        plot1.set_xlabel(xdata[0])
        plot1.set_ylabel(ydata[0])
        plot2 = plt.subplot(222)
        plot2.plot(data[xdata[0]],
                        data[ydata[1]],
                        label = ydata[1],
                        color = colors[1],
                        marker=markers[1])
        plot2.set_title(ydata[1]+' vs '+xdata[0])
        plot2.set_xlabel(xdata[0])
        plot2.set_ylabel(ydata[1])
        plot3 = plt.subplot(223)
        plot3.plot(data[xdata[0]],
                        data[ydata[2]],
                        label = ydata[2],
                        color = colors[2],
                        marker=markers[2])
        plot3.set_title(ydata[2]+' vs '+xdata[0])
        plot3.set_xlabel(xdata[0])
        plot3.set_ylabel(ydata[2])
        plt.tight_layout()
    elif numberofgraphs==3:
        plot1 = plt.subplot(221)
        plot1.plot(data[xdata[0]],
                        data[ydata[0]],
                        label = ydata[0],
                        color = colors[0],
                        marker=markers[0])
        plot1.set_title(ydata[0]+' vs '+xdata[0])
        plot1.set_xlabel(xdata[0])
        plot1.set_ylabel(ydata[0])
        plot2 = plt.subplot(222)
        plot2.plot(data[xdata[0]],
                        data[ydata[1]],
                        label = ydata[1],
                        color = colors[1],
                        marker=markers[1])
        plot2.set_title(ydata[1]+' vs '+xdata[0])
        plot2.set_xlabel(xdata[0])
        plot2.set_ylabel(ydata[1])
        plot3 = plt.subplot(223)
        plot3.plot(data[xdata[0]],
                        data[ydata[2]],
                        label = ydata[2],
                        color = colors[2],
                        marker=markers[2])
        plot3.set_title(ydata[2]+' vs '+xdata[0])
        plot3.set_xlabel(xdata[0])
        plot3.set_ylabel(ydata[2])
        plot4 = plt.subplot(224)
        plot4.plot(data[xdata[0]],
                        data[ydata[3]],
                        label = ydata[3],
                        color = colors[3],
                        marker=markers[3])
        plot4.set_title(ydata[3]+' vs '+xdata[0])
        plot4.set_xlabel(xdata[0])
        plot4.set_ylabel(ydata[3])
        plt.tight_layout()

#twodplot(data, ['T_val'], ['h_val_CEA', 'h_val_pkg'], colors=['red', 'blue', 'green'])
```
